```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import randrange
```

```python
r_rawimg = cv2.imread('right.jpg')
#rawimg1 = cv2.resize(rawimg1, (0,0), fx=1, fy=1)
r_img = cv2.cvtColor(r_rawimg,cv2.COLOR_BGR2GRAY)

l_rawimg = cv2.imread('left.jpg')
#rawimg2 = cv2.resize(rawimg2, (0,0), fx=1, fy=1)
l_img = cv2.cvtColor(l_rawimg,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('Image',r_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(len(r_img))
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(r_img,None)
kp2, des2 = sift.detectAndCompute(l_img,None)
```

```python
cv2.imshow('right.jpg',cv2.drawKeypoints(r_rawimg,kp1,None))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
flann = cv2.FlannBasedMatcher(index_params, search_params)
matches = flann.knnMatch(des1,des2,k=2)
```

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)
```

```python
good = []
for m in matches:
    if m[0].distance < 0.9*m[1].distance:
        good.append(m)
matches = np.asarray(good)        
```

```python
img3 = cv2.drawMatchesKnn(r_rawimg,kp1,l_rawimg,kp2,matches,None)
```

```python
cv2.imshow("output_match.jpg", img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
len(matches)
```

```python
if len(matches[:,0]) >= 4:
    src = np.float32([ kp1[m.queryIdx].pt for m in matches[:,0] ]).reshape(-1,1,2)
    dst = np.float32([ kp2[m.trainIdx].pt for m in matches[:,0] ]).reshape(-1,1,2)
    H, masked = cv2.findHomography(src, dst, cv2.RANSAC, 5.0)
#print H
else:
    raise AssertionError('Canâ€™t find enough keypoints.')
```

```python
dst = cv2.warpPerspective(r_rawimg,H,(l_rawimg.shape[1] + r_rawimg.shape[1], l_rawimg.shape[0]))
plt.subplot(122),plt.imshow(dst),plt.title('Warped Image')
plt.show()
plt.figure()
dst[0:l_rawimg.shape[0], 0:r_rawimg.shape[1]] = l_rawimg
cv2.imwrite('output.jpg',dst)
plt.imshow(dst)
plt.show()
```
